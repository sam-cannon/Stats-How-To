```python
#rename columns
df = df.set_axis(['INTER_40', 'OCCUP_40', 'INTER_60', 'OCCUP_60'], axis=1, inplace=False)
```

```python
#create stress averages for both ages
df['av_40'] = (df['INTER_40'] + df['OCCUP_40'])/2
df['av_60'] = (df['INTER_60'] + df['OCCUP_60'])/2
```

```python
#run paired samples t test between groups
from scipy import stats
stats.ttest_rel(df['av_40'], df['av_60'])
```
